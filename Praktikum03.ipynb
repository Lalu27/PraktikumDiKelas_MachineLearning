# Load Data


```python
import pandas as pd
df = pd.read_csv(path + '/data/socr.csv')
df.head()
```

```python
df.describe()
```

# Pra Proses Data & Konversi data menjadi cm


```python
df1 = (
    df [["Height(Inches)", "Weight(Pounds)"]]
    .rename(columns={"Height(Inches)": "tinggi_cm",
                     "Weight(Pounds)": "berat_kg"})
    .assign(
        tinggi_cm = lambda d: d["tinggi_cm"] * 2.54,
        berat_kg = lambda d: d["berat_kg"] * 0.453359237
    )
    .round({"tinggi_cm" : 2, "berat_kg" : 2})
).copy()
df.head()
```

# Data Pre-Processing

```python
from sklearn.model_selection import train_test_split

X = df1[["tinggi_cm"]]
y = df1["berat_kg"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

# Membagi Data

```python
print ("Jumlah data training", len(X_train))
print ("Jumlah data training", len(y_train))
```

# Pemodelan : Training dan Testing Data

```python
from sklearn.linear_model import LinearRegression

#buat objek instan dari class LinearRegression bernama model
model = LinearRegression()

#Jalankan Training
model.fit(X_train, y_train)
```

# Evaluasi Model


```python
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

#Jalankan prediksi dengan data testing
y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
df_resume = pd.DataFrame()
df_resume['No'] = "1"
df_resume['Koefisien (Kg/cm)'] = model.coef_[0]
df_resume['Intersep (Kg)'] = model.intercept_
df_resume['MAE kg'] = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
df_resume['MSE'] = mean_squared_error(y_test, y_pred)
df_resume
```
